```python
import numpy as np
import matplotlib.pyplot as plt
```


 Dans cet exercice, nous allons créer une variable aléatoire Z=(X,Y) où **X** représentera **la taille** de chaque individus d'une population en centimètres et **Y le poids** de chaque individus de cette population en kilogrammes.

 Pour générer ces valeurs, nous utiliserons une loi normale dont nous choisirons les paramètres.

 Pour X, soit la taille, nous utiliserons une moyenne de 175 cm (qui correspond à la moyenne française) et un écart-type de 5 cm.

 Pour Y, le poids, nous utiliserons une moyenne de 81.2 kg et un écart-type de 10kg.

 Nous générons ensuite nos deux échantillons X et Y de taille n=1000.

```python
mu_x = 175  # Moyenne de la taille (en cm)
sigma_x = 5  # Écart-type de la taille (en cm)

mu_y = 81.2  # Moyenne du poids (en kg)
sigma_y = 10  # Écart-type du poids (en kg)

# Génération des échantillons
n = 1000
x = np.random.normal(mu_x, sigma_x, n)
y = np.random.normal(mu_y, sigma_y, n)
```

On calcule ensuite l'espérance de X,Y et Z.
Dans le cas d'une loi normale, l'espérance de notre variable aléatoire est égale à sa moyenne.

```python
# Esperance de X et Y
x_expected_value = x.mean()
y_expected_value = y.mean()
z_expected_value = (x_expected_value, y_expected_value)

x_expected_value, y_expected_value, z_expected_value
```

Pour la suite, nous allons échantilloner aléatoirement 150 observations de X et Y et les afficher graphiquement sur un nuage de points.

```python
# Taille de l'échantillon
n = 150

# Echantillonage de x et y
x_sample = np.random.choice(x, size = n)
y_sample = np.random.choice(y, size = n)

# Affichage des données
plt.scatter(x_sample, y_sample)
plt.xlabel("Taille (cm)")
plt.ylabel("Poids (kg)")
```


 Afin de vérifier la convergence de la moyenne vers l'espérance, nous calculons la distance euclidienne entre la moyenne et l'espérance pour un échantillon allant de 1 à n + 1.

```python
# Résultats pour chaque augmentation de la taille de l'échantillon
euclidean_distances = []

for i in range (1, 1001) :
  emp_mean = (np.mean(x[:i]), np.mean(y[:i]))
  euclidean_distances.append(np.linalg.norm(np.array(emp_mean) - np.array(z_expected_value)))
```

```python
plt.figure()
plt.plot(range(1, 1001), euclidean_distances)
plt.xlabel("Taille d'échantillon (n)")
plt.ylabel("Distance euclidienne à l'espérance")

plt.show()
```

Nous voyons ici que plus la taille de l'échantillon augmente et plus la distance euclidienne tend vers 0. Nous pounvons donc conclure que la moyenne empirique converge bien vers l'espérance.
